```python
import os
import json
import matplotlib.pyplot as plt
```

```python
with open("logs/base_log_cropped.json", "r") as f:
    base_log_cropped = json.load(f)

with open("logs/base_log.json", "r") as f:
    base_log = json.load(f)

with open("logs/base_log_cross_entropy.json", "r") as f:
    base_log_cross_entropy = json.load(f)

with open("logs/base_log_aug.json", "r") as f:
    base_log_aug = json.load(f)

with open("logs/base_log_aug_cropped.json", "r") as f:
    base_log_aug_cropped = json.load(f)

with open("logs/base_log_adam_di_ce.json", "r") as f:
    base_log_adam_ce = json.load(f)

with open("logs/base_log_adam_di_ce_cropped.json", "r") as f:
    base_log_adam_ce_cropped = json.load(f)
    
with open("logs/base_log_focal.json", "r") as f:
    base_log_focal = json.load(f)

with open("logs/base_log_focal_adam.json", "r") as f:
    base_log_focal_adam = json.load(f)

with open("logs/base_log_focal_cropped.json", "r") as f:
    base_log_focal_cropped = json.load(f)
```

```python
def plot_graphs(json_file):
    plt.figure(figsize=(8, 6))
    plt.plot(json_file["test_iou"], label="Test mIOU")
    plt.plot(json_file["train_iou"], label="Train mIOU")
    plt.plot(json_file["val_iou"], label="Val mIOU")
    
    plt.xlabel('Epochs')
    plt.ylabel('mIOU')
    plt.title('mIOU Plot')
    
    plt.legend()
    
    # Display the plot
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(base_log_cropped["test_iou"], label="base_model_crop")
plt.plot(base_log["test_iou"], label="base_model")
plt.plot(base_log_cross_entropy["test_iou"], label="base_model_cross_entropy")

plt.xlabel('Epochs')
plt.ylabel('mIOU')
plt.title('mIOU Plot for Testing')

plt.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(base_log_cropped["val_iou"], label="base_model_crop")
plt.plot(base_log["val_iou"], label="base_model")
plt.plot(base_log_cross_entropy["val_iou"], label="base_model_cross_entropy")

plt.xlabel('Epochs')
plt.ylabel('mIOU')
plt.title('mIOU Plot for Validaton')

plt.legend()

# Display the plot
plt.show()
```
